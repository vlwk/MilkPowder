```python
import csv
import math
import numpy as np
```

```python
def tf(t, d):
    ret = 0
    for i in d:
        if i == t:
            ret += 1
    return ret
def tf_binary(t, d):
    if t in d:
        return 1
    return 0
def tf_log(t, d):
    return math.log(1 + tf(t, d))
def idf(t, D):
    cnt = 0
    for d in D:
        if t in d:
            cnt += 1
    return math.log(len(D) / cnt)
def idf_smooth(t, D):
    cnt = 0
    for d in D:
        if t in d:
            cnt += 1
    return math.log(1 + len(D) / cnt)
def idf_unary(t, D):
    return 1
def tfidf(t, d, D):
    return tf(t, d) * idf(t, D)
```

```python
documents = []
tags = []
with open("coldstorage4.csv", "r") as f:
    file = csv.reader(f)
    cnt = 0
    itemz=0
    # tmp = 10000
    for line in file:
        if (cnt % 100 == 0 and cnt > 0):
            a = line[7].split(',')
            if(len(a)>=6):
                for tag in a[len(a)-6:]:
                    if tag not in tags:
                        tags.append(tag)
                documents.append(a[len(a)-6:])
                print(cnt,a[len(a)-6:])
                itemz+=1
            # tmp = min(tmp, len(a))
        cnt += 1
        if (itemz >= 100):
            # print(tmp)
            break
tfidfs=[]
for d in documents:
    arr=[]
    for t in d:
        val=tfidf(t,d,documents)
        arr.append(val)
    tfidfs.append(arr)
print(len(tfidfs))
print(tfidfs)
            
```

```python
def cossim(a,b):
    dot=0
    moda=0
    modb=0
    tmp=len(a)
    for i in range(tmp):
        dot+=a[i]*b[i]
    for i in range(tmp):
        moda+=a[i]*a[i]
    moda=math.sqrt(moda)
    for i in range(tmp):
        modb+=b[i]*b[i]
    modb=math.sqrt(modb)
    return dot/(moda*modb)
sim=np.zeros((100,100))
print(sim)
        
for i in range(100):
    for j in range(100):
        if (i != j):
            x=cossim(tfidfs[i],tfidfs[j])
            sim[i,j]=x
print(sim)
```
